### Merging on existing data model output file

```python
final_gdf = gpd.read_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\IDS-DRR-Assam\Maps\Assam_Revenue_Circles\assam_RC_2024-11.geojson')
```

```python
correction_dict = { 'MARIGAON' : "MORIGAON", "SIVSAGAR" : "SIVASAGAR", "K.ANGLONG":"KARBI ANGLONG" }
```

```python
risk_score['district'] = risk_score['district'].replace(correction_dict)
```

```python
risk_score.columns
```

```python
risk_score_rc = risk_score.loc[~risk_score['hq'].isna()]
risk_score_rc
```

```python
final_gdf['object_id_old'] = final_gdf['object_id'].str[-3:].astype(int)
object_id_mapping_rc = final_gdf.set_index('object_id_old')['object_id'].to_dict()
risk_score_rc['object-id'] = risk_score_rc['object-id'].replace(object_id_mapping_rc)
risk_score_rc
```

```python
dist_ids = pd.read_csv(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\IDS-DRR-Assam\RiskScoreModel\assets\district_objectid.csv')
dist_shp = gpd.read_file(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\IDS-DRR-Assam\Maps\as_ids-drr_shapefiles\assam_district_2024-11.geojson')
```

```python
merged = pd.merge(dist_ids, dist_shp[['object_id','dtname']], left_on='district', right_on='dtname',how='right')
merged = merged.drop(['district','object-id'],axis=1).to_csv(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\IDS-DRR-Assam\RiskScoreModel\assets\district_objectid_2.csv',index=False)
```

```python
risk_score_dt = risk_score.loc[risk_score['hq'].isna()]
final_gdf['dtcode11'] = final_gdf['object_id'].str[:6]
dist_ids['district'] = dist_ids['district'].replace(correction_dict)
dist_merge = pd.merge(dist_ids,final_gdf[['dtcode11','dtname']],left_on='district',right_on='dtname')
object_id_mapping_dt = dist_merge.set_index('object-id')['dtcode11'].to_dict()
risk_score_dt['object-id'] = risk_score_dt['object-id'].replace(object_id_mapping_dt)
risk_score_dt
```

```python
final_risk = pd.concat([risk_score_rc, risk_score_dt], ignore_index=True)
final_risk['dtcode11'] = final_risk['object-id'].str[:6]
final_risk
```

```python
district_level = risk_score['rc-area'].isna()  # Identify district-level rows with missing 'object-id'
district_data = risk_score[district_level]

# Perform the merge
merged_district_data = district_data.merge(
    final_gdf[['district', 'dtcode11']],
    on='district',
    how='left'
)

# Step 4: Replace 'object-id' for district-level rows with 'dtcode11'
risk_score.loc[district_level, 'object-id'] = merged_district_data['dtcode11']

# Step 4b: Add the 'dtcode11' column for district-level rows to risk_score
risk_score.loc[district_level, 'dtcode11'] = merged_district_data['dtcode11']

# Step 4c: Ensure all rows in risk_score (district-level and others) have the 'dtcode11' column
if 'dtcode11' not in risk_score.columns:
    risk_score['dtcode11'] = None  # Add the column if it doesn't exist
```
